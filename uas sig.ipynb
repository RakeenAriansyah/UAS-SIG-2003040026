```python
import geopandas as gpd
```

```python
gdf = gpd.read_file("RBI25K_ADMINISTRASI_AR.json")
gdf.head()
# gdf.info()
```

```python
gdf = gdf[['NAMOBJ', 'FCODE', 'SHAPE_AREA', 'SHAPE_LEN', 'geometry']].fillna(0)
```

```python
pip install folium matplotlib mapclassify
```

```python
import folium
m = gdf.explore(column="SHAPE_AREA", name="Shape Area")
folium.LayerControl().add_to(m)
m
```
